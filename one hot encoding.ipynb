```python
import pandas as pd
from sklearn import linear_model
df=pd.read_csv("C:\\Users\\shiny\\Downloads\\homeprices (1).csv")
df
```

```python
#using get_dummies in pandas

dummies=pd.get_dummies(df.town)  # getting categorical variables into dummy variable
dummies
```

```python
merged=pd.concat([df,dummies],axis="columns")
merged
```

```python
final=merged.drop(['town','west windsor'],axis='columns')
final
```

```python
model=linear_model.LinearRegression()
```

```python
#to remove dependent variable 'price'
X=final.drop(['price'],axis='columns') # so training data sets  are 'area','monroe township,'robinsville 

X
```

```python
Y=final.price
Y
```

```python
model.fit(X,Y)
```

```python
model.predict([[1,0,2800]])
```

```python
model.predict([[0,0,3400]])
```

```python
model.score(X,Y)# to find accuracy
```
